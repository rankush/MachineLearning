## Initial Approach
Here we're using logistic regression to detect State of the watch i.e. in motion and stationary

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Let us first plot the data

```python
FILENAME_MOTION = "../logs/motion/1.txt"
FILENAME_STATIONARY = "../logs/stationary/1.txt"

grav_x = []
grav_y = []
grav_z = []
lin_x = []
lin_y = []
lin_z = []

def read_data(filename):
    global grav_x, grav_y, grav_z
    global lin_x, lin_y, lin_z
    text_file = open(filename, "r")
    lines = text_file.readlines()
    text_file.close()
    grav_x = []
    grav_y = []
    grav_z = []
    lin_x = []
    lin_y = []
    lin_z = []
    #print(lines[0])
    for line in lines:
        line_split = line.split(',')
        grav_x.append(float(line_split[0]))
        grav_y.append(float(line_split[1]))
        grav_z.append(float(line_split[2]))
        lin_x.append(float(line_split[3]))
        lin_y.append(float(line_split[4]))
        lin_z.append(float(line_split[5]))
    grav_x = grav_x[25:75]
    grav_y = grav_y[25:75]
    grav_z = grav_z[25:75]
    lin_x = lin_x[25:75]
    lin_y = lin_y[25:75]
    lin_z = lin_z[25:75]
#print(grav_x)
```

Each line is a row of sample data in following format:
[GRAVITY,LINEAR ACCEL].

Let us just plot all the sample against the number of samples

```python
def plotdata():
    plt.subplot(2, 3, 1)
    plt.plot(grav_x)
    plt.yticks(np.arange(-20, 20, step=5))
    plt.xlabel('time (s)')
    plt.ylabel('Gravity[x]')

    plt.subplot(2, 3, 2)
    plt.plot(grav_y)
    plt.yticks(np.arange(-20, 20, step=5))
    plt.xlabel('time (s)')
    plt.ylabel('Gravity[y]')


    plt.subplot(2, 3, 3)
    plt.plot(grav_z)
    plt.yticks(np.arange(-20, 20, step=5))
    plt.xlabel('time (s)')
    plt.ylabel('Gravity[z]')

    plt.subplot(2, 3, 4)
    plt.plot(lin_x)
    plt.yticks(np.arange(-20, 20, step=5))
    plt.xlabel('time (s)')
    plt.ylabel('Linear Accel[x]')

    plt.subplot(2, 3, 5)
    plt.plot(lin_y)
    plt.yticks(np.arange(-20, 20, step=5))
    plt.xlabel('time (s)')
    plt.ylabel('Linear Accel[y]')


    plt.subplot(2, 3, 6)
    plt.plot(lin_z)
    plt.yticks(np.arange(-20, 20, step=5))
    plt.xlabel('time (s)')
    plt.ylabel('Linear Accel[z]')

    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
    plt.show()
```

## Plot Stationary Data

```python
read_data(FILENAME_STATIONARY)
plotdata()
```

## Plot Motion Data

```python
read_data(FILENAME_MOTION)
plotdata()
```
